```python
from sklearn import datasets
digits = datasets.load_digits()
# use the digits dataset
```

```python
# Use PCA to reduce the number of features and preserve the variance
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(digits.data)
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=digits.target, cmap="Paired")
plt.colorbar()
```

```python
# group observations into k groups
iris = datasets.load_iris()
iris_data = iris.data
```

```python
# confirm the number of groups
iris.target_names
```

```python
from sklearn.cluster import KMeans
```

```python
# Use kmeans method to group observations into k=3 groups
k = 3
kmeans = KMeans(n_clusters=k,init = 'k-means++', max_iter = 100, n_init = 10, random_state=42)
y_pred = kmeans.fit_predict(iris_data)
```

```python
print(kmeans.cluster_centers_) 
```

```python
#plotting the centroids of the clusters and visualize
plt.scatter(iris_data[y_pred == 0, 0], iris_data[y_pred == 0, 1],s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(iris_data[y_pred == 1, 0], iris_data[y_pred == 1, 1],s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(iris_data[y_pred == 2, 0], iris_data[y_pred == 2, 1],s = 100, c = 'green', label = 'Iris-virginica')   
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],s = 100, c = 'black', label = 'Centroids')   
plt.legend()
plt.show()
```
